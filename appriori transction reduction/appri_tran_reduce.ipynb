```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('/home/sokhorn/sokhorn/dataSet/data/Online Retail.xlsx')
df  
```

```python
#clear data 
df.Description = df.Description.str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df.InvoiceNo = df.InvoiceNo.astype('str')
df = df[~df['InvoiceNo'].str.contains("C")]
```

```python
basket = (df[df['Country'] == 'United Kingdom']
          .groupby(['InvoiceNo', 'StockCode',])['Quantity'].sum()
          .unstack().reset_index().fillna(0).set_index("InvoiceNo")
          )
basket
```

```python
def end_code(x):
    if(x <= 0):
        return 0
    else:
        return 1
basket =  basket.applymap(end_code)
basket
```

```python
item_set_no_reduse =apriori(basket, min_support=0.07, use_colnames=True)
item_set_no_reduse
```

```python
rc_size = lambda a :  a.groupby(basket.columns.tolist() , as_index=False).size()
```

```python
itemset_rc = rc_size(basket)
itemset_rc
```

```python
itemset_rc =  itemset_rc.loc[:,~itemset_rc.columns.isin(['size'])]
```

```python
item_set_reduse =apriori(itemset_rc, min_support=0.07, use_colnames=True)
item_set_reduse
```

### support count of k item set

```python
min_supp = 1
def support_k_itemset(k_itemset):
    s = 0
    for i in range(k_itemset.shape[0]):
        temp =  int(k_itemset.iloc[i, 0]) 
        for j in range(k_itemset.shape[1] - 1) : 
            temp &= int( k_itemset['size'].loc[i]) & int(k_itemset.iloc[i, j]) 
        s += temp
    return s
```
