```python
sample_dataset = df[[
    'InvoiceNo', 
    'StockCode',
    'Quantity'
]]
sample_dataset
```

```python
item_sets = (
    sample_dataset.groupby(['InvoiceNo', 'StockCode', ])['Quantity']
    .sum().unstack().reset_index().fillna(0)
    .set_index("InvoiceNo")
)
item_sets
```

```python
def en_code(x):
    if x > 0 :
        return 1
    else:
        return 0
```

```python
item_sets = item_sets.applymap(en_code)
item_sets
```
