```python
import pandas as pd
```

```python
df = pd.read_csv('/home/sokhorn/sokhorn/dataSet/data/sample_data_set.csv',sep=',')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df  
```

```python
sample_dataset = df[[
    'InvoiceNo', 
    'StockCode',
    'Quantity'
]]
sample_dataset
```

```python
basket = (
    sample_dataset.groupby(['InvoiceNo', 'StockCode', ])['Quantity']
    .sum().unstack().reset_index().fillna(0)
    .set_index("InvoiceNo")
)
basket
```

```python
def en_code(x):
    if x > 0 :
        return 1
    else:
        return 0
```

```python
dubplicated_items = basket[basket.duplicated()]
dubplicated_items
```

```python
rc_size = lambda a :  a.groupby(dubplicated_items.columns.tolist() , as_index=False).size()
```

```python
basket =  basket.applymap(en_code)
basket
```

```python
rc_df =  rc_size(basket)
rc_df
```

```python
a =  pd.DataFrame(rc_df.iloc[:, rc_df.columns != 'size'].sum(axis=0), columns=['sum']).T
a
```

```python
rc_df =  pd.concat([rc_df, a])
rc_df
```
