# Euclidean distance

```python
def euclidean_distance(x, y):
    return sqrt(sum(pow(x_i-y_i, 2) for x_i, y_i in zip(x, y)))
```

```python
x = [1, 2, 2]
y = [5, 6, 4]
euclidean_distance(x, y)
```

# Manhattan distance

```python
def manhattan_distance(x,y):
    return sum(abs(a-b) for a,b in zip(x,y))
```

```python
x_m = [1, 2, 2]
y_m = [5, 6, 4]
```

```python
manhattan_distance(x_m, y_m)
```

# Minkowski distance

```python
def nth_root(value, r): 
    root_value = 1/float(r)
    return round (Decimal(value) ** Decimal(root_value), 4)

def minkowski_distance(x,y,p_value):
     return nth_root(sum(pow(abs(a-b),p_value) for a,b in zip(x, y)) ,p_value)
```

```python
x_min = [0,2,3]
y_min = [2,0,1]
```

```python
print(minkowski_distance(x_min, y_min, 3))
```

```python
minkowski_distance(x_m, y_m, 3)
```

```python
for i in range(3, 1000):
    print(minkowski_distance(x_min, y_min, i))
```

# Cosine similarity

```python
def square_rooted(x):
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
 
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    return round(numerator/float(denominator),3)
```

```python
x_c = [3, 45, 7, 2]
y_c = [2, 54, 13, 15]
```

```python
cosine_similarity(x_c, y_c)
```

# Jaccard similarity

```python
def jaccard_similarity(x,y):
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality / float(union_cardinality)
```

```python
x_j = [0,1,2,5,6,0, 9]
y_j = [0,2,3,5,7,9, 9]
```

```python
jaccard_similarity(x_j, y_j)
```
