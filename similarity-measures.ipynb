# Euclidean distance

```python
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2) for a, b in zip(x, y)))
```

```python
x = [0,3,4,5]
y = [7,6,3,-1]
```

```python
euclidean_distance(x, y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.xlabel("x", fontsize = 16)
plt.ylabel("y", fontsize=16)
plt.title("Euclidean distance", color = 'blue',fontsize = 'large')
plt.show()
```

# Manhattan distance

```python
def manhattan_distance(x,y):
    return sum(abs(a-b) for a,b in zip(x,y))
```

```python
x_m = [10, 20, 10]
y_m = [10, 20, 20]
```

```python
manhattan_distance(x_m, y_m)
```

```python
plt.plot(x_m, y_m)
plt.xlabel("x", fontsize = 16)
plt.ylabel("y", fontsize=16)
plt.title("Manhattan distance", color = 'blue',fontsize = 'large')
plt.show()
```

# Minkowski distance

```python
def nth_root(value, n_root):
 
    root_value = 1/float(n_root)
    return round (Decimal(value) ** Decimal(root_value),4)
 
def minkowski_distance(x,y,p_value):

    return nth_root(sum(pow(abs(a-b),p_value) for a,b in zip(x, y)),p_value)
 
```

# Cosine similarity

```python
def square_rooted(x):
 
    return round(sqrt(sum([a*a for a in x])),3)
 
def cosine_similarity(x,y):
 
    numerator = sum(a*b for a,b in zip(x,y))
    denominator = square_rooted(x)*square_rooted(y)
    return round(numerator/float(denominator),3)
```

```python
x_c = [3, 45, 7, 2]
y_c = [2, 54, 13, 15]
```

```python
plt.scatter(x_c, y_c)
plt.xlabel("x", fontsize = 16)
plt.ylabel("y", fontsize=16)
plt.title("Cosine distance", color = 'blue',fontsize = 'large')
plt.show()
```

```python
cosine_similarity(x_c, y_c)
```

# Jaccard similarity

```python
def jaccard_similarity(x,y):
    
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
 
```

```python
x_j = [0,1,2,5,6,0]
y_j = [0,2,3,5,7,9]
```

```python
plt.plot(x_j, y_j)
plt.xlabel("x", fontsize = 16)
plt.ylabel("y", fontsize=16)
plt.title("Jaccard distance", color = 'blue',fontsize = 'large')
plt.show()
```

```python
jaccard_similarity(x_j, y_j)
```
